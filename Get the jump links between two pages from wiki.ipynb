```python
import requests 
from requests import request
import re
from bs4 import BeautifulSoup
from urllib.parse import urlparse
from urllib.request import urlopen
```

```python
def get_random_url():
    page = requests.get('https://en.wikipedia.org/wiki/Special:Random')
    return page.url
```

```python
def get_standards_compliant_links(url):
    content=urlopen(url).read()
    BeautS=BeautifulSoup(content,'html.parser')
    return (BeautS.find(id='mw-content-text')).find_all('a',attrs={'href':re.compile('/wiki/')})
```

```python
def get_recursive_links(beginurl,targeturl,depth):
    all_links=get_standards_compliant_links(beginurl)
    if all_links==None:#The last website
        #print(' The  ',biginurl,' ) has no hyperlinks.')
        stop=True
        return stop
    if recursive_path.get(beginurl)==None:
        links_in_page = []#集合
        for hyl in all_links:
            hyperlink=hyl['href']
            
            #href="/wiki/China"
            if not (':' in hyperlink):
                if hyperlink.startswith('/') and not hyperlink.startswith('//'):
                    hyperlink='https://en.wikipedia.org'+hyl['href']
                if urlparse(hyperlink).netloc=='en.wikipedia.org':#network location part is right
                    links_in_page.append(hyperlink)
            if str(hyperlink)==targeturl:
                recursive_path[beginurl]=links_in_page
                stop=True
                return stop
        recursive_path[beginurl]=links_in_page
        #print(recursive_path)
        
        #To recursive
        if not (depth<=0):
            for each in links_in_page:
                stop=get_recursive_links(each,targeturl,depth=depth-1)#If it's not the last, continue
                if stop:
                    break
```

```python
def way_to_target(recursive_path,beginurl,targeturl,path=[]):
    path=path+[beginurl]
    if beginurl==targeturl:#Find the target
        return path
    if recursive_path.get(beginurl)==None:#The page is the last.
        return None
    for each in recursive_path[beginurl]:#To find all
        if each not in path:
            Npath=way_to_target(recursive_path,each,targeturl,path)
            if Npath:
                return Npath
    return None
```
